


# Завдання 1

```python
import numpy as np
import plotly.express as px
import pandas as pd

# Генеруємо випадкові дані для 6 графіків
data1 = np.random.randint(100, 2000, size=100)
data2 = np.random.rand(100) * 1000
data3 = np.random.uniform(low=500, high=1500, size=100)
data4 = np.random.normal(loc=1000, scale=200, size=100)
data5 = np.random.exponential(scale=500, size=100)
data6 = np.random.poisson(lam=1000, size=100)

# Створюємо датафрейм для даних
df = pd.DataFrame({
    'Data1': data1,
    'Data2': data2,
    'Data3': data3,
    'Data4': data4,
    'Data5': data5,
    'Data6': data6
})

# Створюємо графік
fig = px.line(df, x=df.index, y=df.columns, color_discrete_sequence=px.colors.qualitative.Set1)

# Встановлюємо заголовок графіка
fig.update_layout(title_text='6 графіків з різними значеннями')

# Показуємо графік
fig.show()
```

# Завдання 2

```python
import numpy as np
import plotly.express as px
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Генеруємо випадкові дані для 6 графіків
data1 = np.random.randint(100, 2000, size=100)
data2 = np.random.rand(100) * 1000
data3 = np.random.uniform(low=500, high=1500, size=100)
data4 = np.random.normal(loc=1000, scale=200, size=100)
data5 = np.random.exponential(scale=500, size=100)
data6 = np.random.poisson(lam=1000, size=100)

# Створюємо датафрейм для даних
df = pd.DataFrame({
    'Data1': data1,
    'Data2': data2,
    'Data3': data3,
    'Data4': data4,
    'Data5': data5,
    'Data6': data6
})

# Нормалізуємо дані за допомогою MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Створюємо графік
fig = px.line(df_scaled, x=df_scaled.index, y=df_scaled.columns, color_discrete_sequence=px.colors.qualitative.Set1)

# Встановлюємо заголовок графіка
fig.update_layout(title_text='6 графіків з нормалізованими значеннями')

# Показуємо графік
fig.show()
```

```python
import numpy as np
import plotly.express as px

# Генеруємо два розподіли з різними значеннями проміжку
data1 = np.random.normal(loc=500, scale=100, size=1000)
data2 = np.random.normal(loc=1500, scale=200, size=1000)

# Створюємо датафрейм для даних
df1 = pd.DataFrame({'Value': data1})
df2 = pd.DataFrame({'Value': data2})

# Створюємо графік
fig1 = px.histogram(df1, x='Value', nbins=50, title='Розподіл 1')
fig2 = px.histogram(df2, x='Value', nbins=50, title='Розподіл 2')

# Створюємо субплоти
from plotly.subplots import make_subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=['Розподіл 1', 'Розподіл 2'])
fig.add_trace(fig1.data[0], row=1, col=1)
fig.add_trace(fig2.data[0], row=1, col=2)

# Показуємо графік
fig.show()
```

# Завдання 4

```python
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from plotly.subplots import make_subplots

# Генеруємо два розподіли з різними значеннями проміжку
data1 = np.random.normal(loc=500, scale=100, size=1000)
data2 = np.random.normal(loc=1500, scale=200, size=1000)

# Створюємо датафрейм для даних
df1 = pd.DataFrame({'Value': data1})
df2 = pd.DataFrame({'Value': data2})

# Стандартизуємо дані за допомогою StandardScaler
scaler = StandardScaler()
data1_standardized = scaler.fit_transform(df1)
data2_standardized = scaler.fit_transform(df2)

# Створюємо датафрейм для стандартизованих даних
df1_standardized = pd.DataFrame({'Value': data1_standardized.flatten()})
df2_standardized = pd.DataFrame({'Value': data2_standardized.flatten()})

# Створюємо графік
fig1 = px.histogram(df1_standardized, x='Value', nbins=50, title='Стандартизований Розподіл 1')
fig2 = px.histogram(df2_standardized, x='Value', nbins=50, title='Стандартизований Розподіл 2')

# Створюємо субплоти
fig = make_subplots(rows=1, cols=2, subplot_titles=['Стандартизований Розподіл 1', 'Стандартизований Розподіл 2'])
fig.add_trace(fig1.data[0], row=1, col=1)
fig.add_trace(fig2.data[0], row=1, col=2)

# Показуємо графік
fig.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Завантаження даних
df_alarms = pd.read_csv('alarms_no_scalse.csv')

# Перевірка даних
print(df_alarms.head())

# Розділення даних на ознаки та мітки
X = df_alarms.iloc[:, :-1]
y = df_alarms.iloc[:, -1]

# Розділення на навчальну та тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Навчання моделі
model = LinearRegression()
model.fit(X_train, y_train)

# Тестування моделі
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

print(f'Mean Squared Error on original data: {mse}')
```
