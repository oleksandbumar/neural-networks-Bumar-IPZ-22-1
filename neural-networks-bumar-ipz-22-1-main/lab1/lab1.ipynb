# Реалізація простих нейтронних мереж

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models
# Step 1: Read the CSV file
df = pd.read_csv('alarms.csv')
fig = px.scatter_3d (df, x='sound', y='visibility', z='alarm')
fig.show()
# Step 2: Prepare the data
X = df [['sound', 'distance', 'visibility']].values
y = df ['alarm'].values
# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Step 4: Define and compile the neural network model
model = models.Sequential([
])
# layers. Dense (64, activation='relu', input_shape=(X_train.shape[1],)), # layers. Dense (32, activation=' relu'),
layers. Dense (1,input_shape=(X_train.shape[1],), activation='sigmoid')
model.compile(optimizer='adam',
loss='mean_squared_error',
metrics=['mse'])
# Step 5: Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data= (X_test, y_test))
# Step 6: Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python

# Step 7: Use the trained model to make predictions
# Example of making predictions on new data
# ['sound', 'distance', 'visibility']
new_data = np.array([
                      [0.7, 0.1, 0.7],
                      [0.3, 0.7, 0.3],
                      [0.4, 0.5, 0.3],
                      [0.6, 0.5, 0.7],
                      [0.9, 0.1, 0.9],
                      ]) # Replace with your new data
predictions = model.predict(new_data)
print("Predictions:")
print(predictions)
```
